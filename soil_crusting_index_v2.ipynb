```python
import arcpy #charge ArcPy, bibliothèque Python d'ArcGIS
from arcpy.sa import * #importe les fonctions du module Spatial Analyst

arcpy.CheckOutExtension("Spatial") #active l’extension Spatial Analyst pour la session Python.
arcpy.env.overwriteOutput = True #autorise Python à écraser un fichier de sortie existant.
```

```python
LF = Raster("limon_moyenne_pourcentage_90m")
LG = Raster("argile_limon_moyenne_pourcentage_90m")
A = Raster("argile_moyenne_pourcentage_90m")
charge = Raster("CNSW_k_kF_K_L08_raster_reclass")
Corg = Raster("COT_teneurs_2015_2019_L08")

# Conversion du carbone organique en matière organique
MO = (Corg/10) * 1.9

# Formule de base de l'indice de battance
formule_base = ((1.5 * LF) + (0.75 * LG)) / (A + (10 * MO))

# Si la couche de charge existe au pixel, on soustrait 0.2
IB = formule_base - (charge * 0.2)

# Sauvegarde du résultat
IB.save(r"C:\Users\antoine\OneDrive - UCL\MASTER2\Memoire\ArcGIS\Indice-battance-v2\Indice-battance-v2.gdb\indice_battance_v3")
```

```python
#Calcul de la surface de la couche de limon
surface_argile_limon_m2 = float(arcpy.management.GetRasterProperties("argile_limon_moyenne_pourcentage_90m", "COLUMNCOUNT")[0]) * float(arcpy.management.GetRasterProperties("argile_limon_moyenne_pourcentage_90m", "ROWCOUNT")[0]) * float(arcpy.management.GetRasterProperties("argile_limon_moyenne_pourcentage_90m", "CELLSIZEX")[0]) * float(arcpy.management.GetRasterProperties("argile_limon_moyenne_pourcentage_90m", "CELLSIZEY")[0])

#Calcul de la surface de la couche de MO
surface_COT_m2 = float(arcpy.management.GetRasterProperties("COT_teneurs_2015_2019_L08", "COLUMNCOUNT")[0]) * float(arcpy.management.GetRasterProperties("COT_teneurs_2015_2019_L08", "ROWCOUNT")[0]) * float(arcpy.management.GetRasterProperties("COT_teneurs_2015_2019_L08", "CELLSIZEX")[0]) * float(arcpy.management.GetRasterProperties("COT_teneurs_2015_2019_L08", "CELLSIZEY")[0])

print(surface_argile_limon_m2, surface_COT_m2)
```
